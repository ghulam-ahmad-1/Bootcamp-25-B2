Problem Statement:
 Write a function that takes a string and returns the top 3 most frequent words in it (ignoring case and punctuation).
Example:
Input: "The quick brown fox jumps over the lazy dog. The dog barked."
Output: ['the', 'dog', 'quick']  # frequency: the(3), dog(2), quick(1)

```python
Input  = "The quick brown fox jumps over the lazy dog. The dog barked."
x = Input.split(" ")
list = []
for m in x:
    list.append((m.strip(".")).lower())
list
dict = {}
for x in list:
    if x in dict:
        dict[x]+=1
    else:
        dict[x] = 1
dict
a = (sorted(dict.items(), key=lambda item: item[1],reverse=True))
for i in range(3):
    print(a[i])

```

Problem Statement:
 Write a function to flatten a nested list of strings
Example:
input = [ 'nata', 'pizza' [ 'eat', 'tea' ], ['sort'] ]
output [ nata', 'pizza' , â€˜eat', 'tea', 'sort']


```python
input = ["nata","pizza",["eat","tea"],["sort"]]
output = []
for x in input:
    if type(x) is str:
        output.append(x)
    else:
        for y in x:
            output.append(y)
output
```

Problem Statement:
 Write a function that compresses a string using the counts of repeated characters.
 If the compressed string is not smaller than the original, return the original.
Example:
Input: "aaabbccccaaa"
Output: "a3b2c4a3"
Input: "abc"
Output: "abc"

```python
def run_len(string1):
    list = []
    prev_letter = ""
    for idx,letter in enumerate(string1):
        if letter != prev_letter:
            current_letter = letter
            count = 0
            while string1[idx+count:idx+count+1] == current_letter:
                count+=1
            prev_letter = current_letter
            list.append(str(count))
            list.append(current_letter)
    xyz = "".join(list)
    if len(xyz) == 2*len(string1):
        return string1  
    else:
        return xyz
        

print(run_len("aaabcddd"))
print(run_len("abcd"))
```
